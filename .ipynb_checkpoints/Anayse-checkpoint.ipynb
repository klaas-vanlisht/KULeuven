```python
import glob
import astropy.io.fits as pf

import numpy as np
from scipy import interpolate
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

from astropy import constants as const
```

```python
sternaam = 'BD-11_162'

# Maak een lijst van alle spectra. Data is in FITS format
datalist = glob.glob("%s/*.fits" %sternaam)
datalist.sort()
#print(datalist)


# Inlezen van de spectrale lijnen. De goflengte is in rust en in lucht (niet vacuum)
inputfile = "%s/LijnLijst.txt" %sternaam
lines,weights = np.loadtxt(inputfile,usecols=(0,1),unpack=True)
```

```python
# Neem een willekeurig spectrum als voorbeeld
# Spec = intensiteit
# Header laadt de metadata in (o.a. wie de waarnemer was, datum van de observatie...)
spec = pf.getdata(datalist[2])
header = pf.getheader(datalist[2])
```

```python
def golflengte_grid(spec, header):

    # Lees meta-data
    ref_pix = int(header['CRPIX1'])-1   # index van de referentiepixel
    ref_val = float(header['CRVAL1'])   # ln(golflengte) van de referentiepixel
    ref_del = float(header['CDELT1'])   # breedte van de pixel in eenheid van CRVAL1
    JD = header['BJD']  # datum van de waarneming in Juliaanse dagen
    unitx = header['CTYPE1']

    numberpoints = spec.shape[0]

    # Maak een golflengtegrid aan
    wavelengthbegin = ref_val - ref_pix*ref_del
    wavelengthend = wavelengthbegin + (numberpoints-1)*ref_del
    wavelengths = np.linspace(wavelengthbegin,wavelengthend,numberpoints)
    wavelengths = np.exp(wavelengths)

    return wavelengths, JD, unitx


wavelengths, JD, unitx = golflengte_grid(spec, header)
```

```python
figalle, allg = plt.subplots()


# begin, einde = 1000, 1650
# llg.plot(wavelengths[begin:einde], spec[begin:einde])
allg.plot(wavelengths, spec)

allg.set_ylabel("intensiteit (flux)")
allg.set_xlabel(r"golflengte")
plt.tight_layout()
plt.show()
#plt.savefig('Spectrum.png')
```

```python
llfig, llg = plt.subplots()
llg.plot(lines, weights)

llg.set_ylabel("intensiteit (flux)")
llg.set_xlabel("golflengte")
plt.tight_layout()
plt.show()
```

# Radiale Snelheid bepalen

```python
# willekeurige data
file = "00307959_HRF_OBJ_ext_CosmicsRemoved_log_merged_cf"
vbfile = "%s/00307959_HRF_OBJ_ext_CosmicsRemoved_log_merged_cf.fits" %sternaam

vbdatalist = glob.glob(vbfile)
vbdatalist.sort()
vbspec = pf.getdata(vbdatalist[0])
vbheader = pf.getheader(vbdatalist[0])

vbwl, vbJD, vbunitx = golflengte_grid(vbspec, vbheader)
# vbwl
```

```python
vbfig, vbllg = plt.subplots()

vbllg.plot(vbwl, vbspec)
plt.show()
```

```python
# functie definieren voor radiale snelheid
c = const.c * 10**-3

def doppler(v, lambda0):
    return (v*lambda0)/c + lambda0 # (lambda - lambda0)/ lambda0 = v/c
```

```python
# vind de best passende snelheid volgends de doppler effect
def beste_v(wavelength, spec, lines, v_bereik, v_steps):
    # genereer snelheden om te testen
    v_test = np.arange(v_bereik[0], v_bereik[1] + v_steps, v_steps)
    # maak interpolatiefunctie
    interpolate_functie = interpolate.interp1d(wavelength, spec)
    return interpolate_functie
r = beste_v(vbwl, vbspec, lines, (-10, 10), 0.1)
print(r(4000))
```

```python
t = np.arange(4000, 6000, 0.1)
y = r(t)

fig, ax = plt.subplots()
# ax.scatter()
```

```python
# volgens Gemini
def find_best_vr(vbwl, vb_spec, lines, v_range_kms, v_step_kms=0.1):
    v_test = np.arange(v_range_kms[0], v_range_kms[1] + v_step_kms, v_step_kms)
    min_flux_sum = []
    
    # Maak een interpolatiefunctie van de waargenomen data
    interp_func = interpolate.interp1d(vbwl, vb_spec)
    
    wl_min, wl_max = vbwl.min(), vbwl.max()
    
    for v_r in v_test:
        # Verschuiving: Gebruikt de logica van jouw Doppler formule: lambda = (v_r * lambda0) / c + lambda0
        lambda_shifted = (v_r * lines) / c + lines
        
        # Filter alleen de lijnen die binnen het waargenomen bereik vallen
        valid_shifted_lines = lambda_shifted[(lambda_shifted >= wl_min) & (lambda_shifted <= wl_max)]
        
        if valid_shifted_lines.size > 0:
            # Minimalisatie: Zoek de flux op de verschoven posities en tel op
            flux_at_shifted_lines = interp_func(valid_shifted_lines)
            min_flux_sum.append(flux_at_shifted_lines.sum())
        else:
            min_flux_sum.append(np.inf)
            
    min_flux_sum = np.array(min_flux_sum)
    best_v_r_index = np.argmin(min_flux_sum)
    best_v_r = v_test[best_v_r_index]
    
    return best_v_r


a = find_best_vr(vbwl, vbspec, lines, (-50, 50), 0.1)
print(f"{round(a, 2)}km per s is de best passende snelheid")
```
