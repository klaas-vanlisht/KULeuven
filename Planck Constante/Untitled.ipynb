```python
ps = [1.70, 1.71, 1.73, 1.84]
for _ in range(12):
    ps.append(round(ps[len(ps) - 1] + 0.1, 3))
```

```python
resistorv = [0, 0, 1, 5, 10, 15.5, 22, 28, 34, 40.5, 47, 53, 60, 68, 75, 84]
resistora = [2, 2, 4, 22, 46, 71.5, 98, 126, 155, 184.5, 213, 242, 271, 307, 341, 382]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(resistorv, resistora)
```

```python
plt.plot(ps, resistorv)
plt.plot(ps, resistora)
```

```python
led = []
for i in range(len(ps)):
    led.append(ps[i] - resistorv[i]/100)
```

```python
x = np.linspace(0, 80, 100)
y = 4.5*x+1

plt.plot(x, y)

plt.plot(led, resistora)
```
